```python
class Person:
#Parent class
    def __init__(self,fname,lname, number= None):
        self.fname = fname
        self.lname = lname
        self.number = number
    def printname(self):
        print(self.lname, self.fname,)

#child class
class Students(Person):
    pass


class Teachers(Students):
    def printnumbers(self):      
        print(self.number, self.lname, self.fname)

        
s1 = Students("Harivignesh", "D")
s1.printname()

t1 = Teachers("Sridevi", "M", 2004)
t1.printnumbers()
```
